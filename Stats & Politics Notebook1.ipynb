## Part 3: Visualizations

### Produce 5 unique and meaningful visualizations from this dataset.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

### Visual 1

```python
plt.figure(figsize=(20,10))
plt.bar(sp['State'], sp['Votes'])
plt.xlabel('States',size=14)
plt.ylabel('Number of Votes',size=14)
plt.title('Total number of votes per state',size=18)
plt.xticks(rotation='vertical')
plt.show()
```

Visual one shows the total number of votes per state. This is helpful, especially in an electoral college voting system, as certain states hold more weight in terms of number of votes. For political campaigns, campaining efforts should be more focused towards the states where a larger number of citizens are more likely to vote. In this case, California and Illinois have almost or more than double the number of votes in comparison to states like Arizona or Florida. Therefore, campaigning efforts in California and Illinois would prove to be more useful.  

### Visual 2

```python
fig=plt.figure(figsize=(20,10))
plt.bar(sp['State'], sp['Violent.crime'])
plt.xlabel('States',size=14)
plt.ylabel('Violent crime rate',size=14)
plt.title('Total violent crime rate per state',size=18)
plt.xticks(rotation='vertical')
plt.show()
```

Visual two shows the total violent crime rate per state. This is helpful in determining which states are in more need of police and crime reform. This information can be used when political campaigns travel across the country to acknowledge the needs of individual states. For example, Arkansas and Missouri have the highest crime rates. The citizens of these states are therefore more likely to vote for the candidate that emphasizes effective police and crime reform. Also, political campaigns can potentially use this data to not only determine which states have a lower crime rate, such as Vermont and North Dakota, but use the safety models from lower crime rate states to determine what reform should look like in the states with a higher crime rate. 

### Visual 3

```python
fig = plt.figure(figsize=(20,70))
sp.iloc[0:20].plot(y=['Less Than High School Diploma','Less Than High School Diploma',"At Least Bachelors's Degree",'Graduate Degree','School Enrollment'],kind='bar',figsize=(20,10))
plt.legend(loc = 'upper right')
plt.xlabel('Counties',size=14)
plt.ylabel('Levels of education',size=14)
plt.title('Levels of education according to county',size=18);
```

Visual three shows the levels of education for the first 19 county entries in the data set. This visual is helpful, as it can be adjusted to show the education levels of any of the counties in the data set, isolated according to state, proximity, population size and a variety of other filters. For the purposes of this assignment, I've only displayed the first 19 counties listed in the data for ease of reference. This visualization is helpful in determining which states have higher levels of education. It can be assumed that since education in the United States is more expensive, that the states with higher education levels are more likely to have a larger number of citizens with student loan debt. Therefore, when planning political campaigns, it would be helpful for strategists to emphasize effective ways to pay off debt and student loans, easing interest rate on student loan payments and making education less expensive, while campaigning in the states with higher education levels. 

### Visual 4

```python
fig = plt.figure(figsize=(20,10))
plt.plot(sp.index, sp['White'], label='White', marker='.',color='red', linewidth=1.4)
plt.plot(sp.index, sp['Black'], label='Black', marker='o', color='black', linewidth=1.5)
plt.plot(sp.index, sp['Hispanic'], label='Hispanic', marker='D', color='blue', linewidth=2)


plt.xlabel('County', size=12)
plt.ylabel('Percentage of people', size=12)
plt.legend()
plt.title('Percentage of citizens by race according to county', size=18)
plt.grid(b=True)
plt.xlim(0,49)
plt.show()
```

Visual four shows the percentage of people by race for the first 50 counties in the data set. This visual can be used to direct campaigns towards race related issues such as Black Lives Matter. For counties with a higher number of Black citizens, political messaging can be targeted towards support for Black Lives Matter. It can also be seen from the data that a majority of the first 50 counties have a much higher White population in comparison to the other races. From this, it can be assumed that these counties are more likely to vote for a White candidate than a candidate who is coloured. 

### Visual 5

```python
plt.figure(figsize=(20,10))
plt.scatter(sp['State'], sp['Poverty.Rate.below.federal.poverty.threshold'],c='red' ,label = "Family poverty rate")
plt.scatter(sp['State'], sp['Child.Poverty.living.in.families.below.the.poverty.line'],c='blue' ,label = "Child poverty rate")
plt.legend()
plt.xlabel('States', size=12)
plt.ylabel('Percentage of poverty',size=18)
plt.title('Percentage of family and child poverty rate per state')
plt.xticks(rotation='vertical')
plt.show();
```

Visual five shows the percentage of family and child poverty rate per state. This can be helpful in determinig the country's overall economic condition and planning ways to provide more employment opportunities, better housing conditions, social services and community resources for states with a higher poverty rate. States with a higher poverty rate are also less likely to vote in elections as they don't have the tools or resources to be registered to vote or be informed about how to make an educated vote.
